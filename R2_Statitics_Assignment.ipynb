```python
#taking random samples from male and female for  BMI
random_female_data = female_data['bmi'].sample(n=600, random_state=1)
```

```python
random_male_data = male_data['bmi'].sample(n=600, random_state=1)
```

```python
#Ho = > male and female bmi is same 
#HA -= > male and female bmi is not  same 
#sample size n =600
# two samples
# Here we select  ð›¼  = 0.05 and sample size = 600 
from scipy import stats 
t_statistic, p_value  =  stats.ttest_rel(random_female_data, random_male_data )
print('P Value %1.3f' % p_value)
#0.085 > 0.05 which is p value is less than 0.05
#So the statistical decision is not  to reject the null hypothesis at 5% level of significance.

#So there is sufficient evidence to accept  hypothesis that  There is no significant difference between BMIS for male and female.
```

```python
# Is the distribution of bmi across women with no children, one child and two children, the same ?
female_no_children_data = data.loc[data['children'] == 0]
female_one_children_data = data.loc[data['children'] == 1]
female_two_children_data = data.loc[data['children'] == 2]
female_three_children_data = data.loc[data['children'] == 3]
```

```python
print(female_no_children_data.shape)
print(female_one_children_data.shape)
print(female_two_children_data.shape)
print(female_three_children_data.shape)
```

```python
# taking random 150 samples for each category
random_female_no_Children_data = female_no_children_data['bmi'].sample(n=150, random_state=1)
random_female_one_Children_data = female_one_children_data['bmi'].sample(n=150, random_state=1)
random_female_two_Children_data = female_two_children_data['bmi'].sample(n=150, random_state=1)
random_female_three_Children_data = female_three_children_data['bmi'].sample(n=150, random_state=1)
```

```python
print('Count, Mean and standard deviation  of random_female_no_Children_data : %3d, %3.2f and %3.2f' % (len(random_female_no_Children_data), random_female_no_Children_data.mean(),np.std(random_female_no_Children_data,ddof =1)))
print('Count, Mean and standard deviation  of random_female_one_Children_data : %3d, %3.2f and %3.2f' % (len(random_female_one_Children_data), random_female_one_Children_data.mean(),np.std(random_female_one_Children_data,ddof =1)))
print('Count, Mean and standard deviation  of random_female_two_Children_data : %3d, %3.2f and %3.2f' % (len(random_female_two_Children_data), random_female_two_Children_data.mean(),np.std(random_female_two_Children_data,ddof =1)))
print('Count, Mean and standard deviation  of random_female_no_Children_data : %3d, %3.2f and %3.2f' % (len(random_female_three_Children_data), random_female_three_Children_data.mean(),np.std(random_female_three_Children_data,ddof =1)))
```

```python
df1            = pd.DataFrame({'sample': '1', 'data':random_female_no_Children_data})
df2            = pd.DataFrame({'sample': '2', 'data':random_female_one_Children_data})
df3            = pd.DataFrame({'sample': '3', 'data':random_female_two_Children_data})
df4            = pd.DataFrame({'sample': '3', 'data':random_female_three_Children_data})

sample_inc_df = pd.DataFrame()
sample_inc_df = sample_inc_df.append(df1) 
sample_inc_df = sample_inc_df.append(df2) 
sample_inc_df = sample_inc_df.append(df3)
sample_inc_df = sample_inc_df.append(df4)


sns.boxplot(x = "sample", y = "data", data = sample_inc_df)
plt.title('BMI  data')
plt.show()


#ð»0 :  ðœ‡1  =  ðœ‡2  =  ðœ‡3 =U4
#ð»ð´ : At least one  ðœ‡  differs

import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
 
crit = stats.f.ppf(q = 1-0.05, dfn = 2, dfd = 60)
print('F critical value for 2 and 60 df with .95 confidence %3.2f' %crit)

mod = ols('data ~ sample', data = sample_inc_df).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

```python
#calculated value of F ( = 0.446906) is very  less than Critical value of F( = 3.15)
#So the statistical decision is to accept  the null hypothesis at 5% level of significance.
#So there is sufficient evidence to accept the null hypothesis that at almost bmi for each categories are same.
```
